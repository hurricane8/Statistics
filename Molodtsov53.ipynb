```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
import matplotlib.gridspec as gridspec
```

```python
fg = plt.figure(figsize=(12, 9), constrained_layout=True, facecolor = "whitesmoke")
gs = gridspec.GridSpec(ncols=1, nrows=1, figure=fg)
x=[0.99, 0.95, 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1, 0.05, 0.01, 0] 
y=[0.0011, 0.0056, 0.0112, 0.0228, 0.0351, 0.0488, 0.0644, 0.0830, 0.1066, 0.1394, 0.19468, 0.2488, 0.368, 1] # данные для проверки по функции y=1/x
fig_ax_1 = fg.add_subplot(gs[0, 0], facecolor = "white")

fig_ax_1.plot(x, y, 'black', linewidth=5)
fig_ax_1.legend(loc="upper right", fontsize=15)
fig_ax_1.set_xlabel(r"$\alpha$", fontsize=15)
fig_ax_1.set_ylabel(r"$\beta$", fontsize=15)
fig_ax_1.set_title(r'$\beta(\alpha)$', fontsize=15)
fig_ax_1.grid(color='darkred')
```
